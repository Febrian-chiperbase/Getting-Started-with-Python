```python
import numpy as np
import matplotlib.pyplot as plt


f = lambda x, y: x**2 - 4*y
h = 0.1
x = np.arange(0, 1 + h, h)
y0 = 1

y_approx = np.zeros(len(x))
y_approx[0] = y0

for i in range(0, len(x) - 1):
    y_approx[i + 1] = y_approx[i] + h * f(x[i], y_approx[i])


def y_exact_func(x_val):
    return (x_val**2)/4 - x_val/8 + 1/32 + (31/32)*np.exp(-4*x_val)

y_exact = y_exact_func(x)


plt.style.use('seaborn-v0_8-poster')
plt.figure(figsize=(12, 8))
plt.plot(x, y_approx, 'bo--', label='Approximate (Euler Method)')
plt.plot(x, y_exact, 'g-', label='Exact Solution')
plt.title('Approximate and Exact Solution for $y\' + 4y = x^2$')
plt.xlabel('x')
plt.ylabel('y(x)')
plt.grid(True)
plt.legend(loc='upper right')
plt.show()


print("x | Euler Approx. | Exact Solution | Error")
print("---------------------------------------------------")
for i in range(len(x)):
    error = abs(y_exact[i] - y_approx[i])
    print(f"{x[i]:.1f} | {y_approx[i]:.6f}    | {y_exact[i]:.6f}     | {error:.6f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt


f = lambda x, y: np.sin(y)
h = 0.1  # Step size
x = np.arange(0, 2 + h, h)
y0 = 1

y_approx = np.zeros(len(x))
y_approx[0] = y0

for i in range(0, len(x) - 1):
    y_approx[i + 1] = y_approx[i] + h * f(x[i], y_approx[i])

tan_half_y0 = np.tan(1.0/2.0)
def y_exact_func(x_val):
    return 2 * np.arctan(tan_half_y0 * np.exp(x_val))

y_exact = y_exact_func(x)

plt.style.use('seaborn-v0_8-poster')
plt.figure(figsize=(12, 8))
plt.plot(x, y_approx, 'bo--', label='Approximate (Euler Method)')
plt.plot(x, y_exact, 'g-', label='Exact Solution')
plt.title('Approximate and Exact Solution for $y\' = \sin(y)$, $y(0)=1$')
plt.xlabel('x')
plt.ylabel('y(x) (radians)')
plt.grid(True)
plt.legend(loc='lower right')
plt.show()


print("x   | Euler Approx. | Exact Solution | Error")
print("---------------------------------------------------")
for i in range(len(x)):
    if i % (len(x)//10) == 0 or i == len(x)-1 :
        error = abs(y_exact[i] - y_approx[i])
        print(f"{x[i]:.1f} | {y_approx[i]:.6f}    | {y_exact[i]:.6f}     | {error:.6f}")
```
