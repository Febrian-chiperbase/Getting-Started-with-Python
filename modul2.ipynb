```python
import numpy as np
import matplotlib.pyplot as plt

# Soal 1
def gauss_seidel_1(x, y, z, tol=0.001, max_iter=1000):
    for _ in range(max_iter):
        x_new = (10 - 3*y + z) / 2
        y_new = (10 - 2*x_new + z) / 3
        z_new = (4 - x_new + y_new) / 2

        if abs(x_new - x) < tol and abs(y_new - y) < tol and abs(z_new - z) < tol:
            return x_new, y_new, z_new

        x, y, z = x_new, y_new, z_new

    return x, y, z

# Initial values
x1, y1, z1 = 5, 5, 5
solution_1 = gauss_seidel_1(x1, y1, z1)
print(f"Solution Soal 1 (Gauss-Seidel): x = {solution_1[0]:.6f}, y = {solution_1[1]:.6f}, z = {solution_1[2]:.6f}")

A1 = np.array([[2, 3, -1], [1, -1, 2], [3, 2, 1]])
B1 = np.array([10, 4, 14])
solution_np_1 = np.linalg.solve(A1, B1)
print(f"Solution Soal 1 (NumPy): x = {solution_np_1[0]:.6f}, y = {solution_np_1[1]:.6f}, z = {solution_np_1[2]:.6f}\n")
```

```python
# Soal 2
def gauss_seidel_2(x, y, z, tol=0.001, max_iter=1000):
    for _ in range(max_iter):
        x_new = (7.85 + 0.1*y + 0.2*z) / 3
        y_new = (-19.3 - 0.1*x_new + 0.3*z) / 7
        z_new = (71.4 - 0.3*x_new + 0.2*y_new) / 10

        if abs(x_new - x) < tol and abs(y_new - y) < tol and abs(z_new - z) < tol:
            return x_new, y_new, z_new

        x, y, z = x_new, y_new, z_new

    return x, y, z

x2, y2, z2 = 1, 2, 3
solution_2 = gauss_seidel_2(x2, y2, z2)
print(f"Solution Soal 2 (Gauss-Seidel): x = {solution_2[0]:.6f}, y = {solution_2[1]:.6f}, z = {solution_2[2]:.6f}")

# Using NumPy
A2 = np.array([[3, -0.1, -0.2], [0.1, 7, -0.3], [0.3, -0.2, 10]])
B2 = np.array([7.85, -19.3, 71.4])
solution_np_2 = np.linalg.solve(A2, B2)
print(f"Solution Soal 2 (NumPy): x = {solution_np_2[0]:.6f}, y = {solution_np_2[1]:.6f}, z = {solution_np_2[2]:.6f}\n")
```

```python
# Soal 3
X = np.array([5, 4, 6, 7, 5])
Y = np.array([10, 8, 12, 14, 10])

# Least Square Method
n = len(X)
sum_X = np.sum(X)
sum_Y = np.sum(Y)
sum_XY = np.sum(X * Y)
sum_X2 = np.sum(X**2)

a = (n * sum_XY - sum_X * sum_Y) / (n * sum_X2 - sum_X**2)
b = (sum_Y - a * sum_X) / n

print(f"Solution Soal 3 (Least Square Method): a = {a:.6f}, b = {b:.6f}")

# Plotting
plt.scatter(X, Y, color='red', label='Data Points')
plt.plot(X, a*X + b, color='blue', label='Regression Line')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Least Square Regression')
plt.legend()
plt.grid(True)
plt.show()
```
